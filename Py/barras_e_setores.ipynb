```python
import pandas as pd
```

```python
df = pd.read_csv('data/insect.csv')
df.shape
```

```python
df.head()
```

```python
#Agrupamento baseado no atributo 'spray'
agrup = df.groupby(['spray'])['count'].sum()
agrup
```

```python
agrup.plot.bar(color= 'gray')
```

```python
agrup.plot.bar(color= ['gray', 'blue', 'yellow', 'red', 'green', 'orange'])
```

```python
#Grafico de setores
agrup.plot.pie()
```

```python
agrup.plot.pie(legend = True)
```
