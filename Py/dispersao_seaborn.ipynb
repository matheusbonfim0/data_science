```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
df = pd.read_csv('data/co2.csv')
df.head()
```

```python
srn.scatterplot(x=df.conc, y=df.uptake, hue=df.Type)
```

```python
#Seleção dos registros específicos da base de dados (Quebec e Mississippi)
q = df.loc[df['Type'] == 'Quebec']
m = df.loc[df['Type'] == 'Mississippi']
```

```python
#Montando subgraficos
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=q.conc, y=q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(x=m.conc, y=m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
#Conferindo dados chilled e nonchilled
ch = df.loc[df['Treatment'] == 'chilled']
nc = df.loc[df['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=ch.conc, y=ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
srn.scatterplot(x=nc.conc, y=nc.uptake).set_title('Non chilled')
plt.tight_layout()
```
