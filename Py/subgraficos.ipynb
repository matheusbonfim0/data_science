```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/trees.csv')
df.head()
```

```python
#Perimetro com volume
plt.scatter(df.Girth, df.Volume)
```

```python
#Perimetro com altura
plt.scatter(df.Girth, df.Height)
```

```python
#Altura com volume
plt.scatter(df.Height, df.Volume)
```

```python
#Histograma volume
plt.hist(df.Volume)
```

```python
#Imprimindo grficos juntos
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(df.Girth, df.Volume)
plt.subplot(2,2,2)
plt.scatter(df.Girth, df.Height)
plt.subplot(2,2,3)
plt.scatter(df.Height, df.Volume, marker='^')
plt.subplot(2,2,4)
plt.hist(df.Volume)
```
