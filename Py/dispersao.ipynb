```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/trees.csv')
df.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(df.Girth, df.Volume, color = 'blue',
facecolors = 'None', marker='*')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
#Gráfico de linha considerando o volume e a dispersão
plt.plot(df.Girth, df.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
#Gráfico de dispersão com 'afastamento' dos dados (jitter)
sns.regplot(x=df.Girth, y=df.Volume, data=df, x_jitter=0.3, fit_reg= False)
```
